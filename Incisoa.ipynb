# Ecuacion de Laplace inciso a

Para resolver la ecuacion de laplace:
$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}=0,  0<x<1, 0<y<1$$
Para las condiciones de frontera:
$$u(0,y)=u(1,y)=0;u(x,0)=100;u(x,1)=200$$


```python
from pylab import *
```

Graficando condiciones de frontera:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,200)

x1 = np.linspace(0, 1,200)
y = np.linspace(0, 1,200)

plot(x1, [100 for x in x1], color="g",label='$f(x)$ = 100, x en $[0, 1]$')
plot(y, [200 for x in y], color="yellow",label='$g(x)$ = 200, x en $[0, 1]$')

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("Condiciónes de frontera (u(x,1)=f(x) y u(x,1)=g(x))"),xlim([-0.01,1.01]),ylim([0.0,225.2]),grid(),legend()
show()
```

Graficando aproximaciones de condiciones de frontera:

```python
def u_0(x,N=10): 
    f = zeros(len(x))
    for n in range(1,N):
        f += ((1-(-1)**n)/n)*sin(n*pi*x)
    return (200.0/pi)*f
f_n = u_0(x)
plot(x,f_n,label='$f_N(x)$ con $N=10$',color="yellow")
x1 = np.linspace(0, 1,200)

plot(x1, [100 for x in x1], color="g",label='$f(x)$ = 100, x en $[0, 1]$')
plt.ylabel('$f(x)$'); plt.xlabel('$x$'); 
plt.title("Aproximación de la condición de frontera u(x,0)=100")
xlim([-0.01,1.01]),ylim([0.0,125.2]),grid(),legend()
show()
```

```python
def u_1(x,N=10): 
    f = zeros(len(x))
    for n in range(1,N):
        f += ((1-(-1)**n)/n)*sin(n*pi*x)
    return (400.0/pi)*f
g_n = u_1(x)
plot(x,g_n,label='$g_N(x)$ con $N=10$', color="yellow")
x1 = np.linspace(0, 1,200)

plot(x1, [200 for x in x1], 'g',label='$g(x)$ = 200, x en $[0, 1]$')
plt.ylabel('$g(x)$'); plt.xlabel('$x$'); 
plt.title("Aproximación de la condición de frontera u(x,1)=200")
xlim([-0.01,1.01]),ylim([0.0,255.2]),grid(),legend()
show()
```

Graficando Solucion 3D:

```python

def u_x_y(x,y):
    N = 5
    f = zeros ((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          a = (200*(1+(-1)^(n+1)))/(n*pi)
          b = (200*(1+(-1)^(n+1))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,i] += sin(n*pi*x)*(a*cosh(n*pi*y[i])+b*sinh(n*pi*y[i]))
    return f
```

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="inferno"); #Otros colores:"viridis", "inferno", "plasma"
```
